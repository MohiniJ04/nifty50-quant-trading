```python
import pandas as pd
import numpy as np
```

```python
file_path = "../data/nifty_cleaned_5min.csv"

df = pd.read_csv(file_path, parse_dates=["datetime"])
df.set_index("datetime", inplace=True)

df.head()
```

```python
df["ema_5"] = df["close"].ewm(span=5, adjust=False).mean()
df["ema_15"] = df["close"].ewm(span=15, adjust=False).mean()
```

```python
df["ema_gap"] = df["ema_5"] - df["ema_15"]
```

```python
df["returns"] = df["close"].pct_change()
```

```python
df["volatility"] = df["returns"].rolling(window=20).std()
```

```python
df["candle_range"] = df["high"] - df["low"]
df["candle_body"] = (df["close"] - df["open"]).abs()
```

```python
df["hour"] = df.index.hour
df["minute"] = df.index.minute
df["day_of_week"] = df.index.dayofweek
```

```python
df["returns_lag_1"] = df["returns"].shift(1)
df["returns_lag_3"] = df["returns"].shift(3)
df["ema_gap_lag_1"] = df["ema_gap"].shift(1)
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
output_path = "../data/nifty_features_5min.csv"
df.to_csv(output_path)

print(f"Feature dataset saved to {output_path}")
```
